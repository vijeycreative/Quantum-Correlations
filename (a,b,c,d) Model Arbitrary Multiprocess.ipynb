```python
import math
import itertools
import numpy as np
import multiprocessing
from multiprocessing import Process, Pool, Array
from tqdm import tqdm
import time
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
def probDist(a,b,c):
    d = 1 - (a+b+c)
    pArr = np.array([[a,b,c,d],
                 [b,c,d,a],
                 [c,d,a,b],
                 [d,a,b,c]])
    return pArr
```

```python
def H(a):
    if(a == 0):
        return 0
    else:
        return (a*math.log2(1/a))
```

```python
def mutualInfo(a,b,c):
    d = 1 - (a+b+c)
    return 2 - H(a) - H(b) - H(c) - H(d)
```

```python
def calculateViolation(arr):
    x = np.arange(0.0, 1.01, 0.01)
    y = np.arange(0.0, 1.01, 0.01)
    for a in x:
        for b in x:
            for c in y:
                if(a + b + c <= 1.00):
                    pArr = probDist(a,b,c)
                    xy = 0
                    xY = 0
                    Xy = 0
                    XY = 0
                    for i in range(0, arr.shape[1]):
                        xy += arr[i,0]*arr[i,2]*pArr[i,0]
                        xY += arr[i,0]*arr[i,3]*pArr[i,1]
                        Xy += arr[i,1]*arr[i,2]*pArr[i,2]
                        XY += arr[i,1]*arr[i,3]*pArr[i,3]
                    E = xy + xY +Xy - XY
                    MI = mutualInfo(a,b,c)
                    if(abs(E) >= 3.9 and MI <= 0.5):
                        print("a = ", a, " b = ", b, " c = ", c, " MI = ", MI, " E = ", abs(E))
                        #print(arr)
                        #print(" ")
```

```python
values = [-1,1]
seq = itertools.product(values, repeat=16)
settingArrays = []
for s in seq:
    # convert to numpy array and reshape to 4x4
    settingArray = np.fromiter(s, np.int8).reshape(4, 4)
    settingArrays.append(settingArray)
```

```python
p = Pool(multiprocessing.cpu_count())
p.map(calculateViolation, settingArrays)
```
